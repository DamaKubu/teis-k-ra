```python
from PIL import Image
import pytesseract

# Function to perform OCR
def ocr_lithuanian(image_path):
    """
    This function takes the path of an image file and performs OCR,
    attempting to interpret the text content of the image using the
    Lithuanian language data for Tesseract.
    """
    try:
        # Open the image file
        image = Image.open(image_path)
        # Specify the path for tesseract executable if it's not in the PATH
        # pytesseract.pytesseract.tesseract_cmd = r'<full_path_to_your_tesseract_executable>'

        # Perform OCR using the Lithuanian language
        ocr_result = pytesseract.image_to_string(image, lang='lit')

        return ocr_result
    except Exception as e:
        print(f"An error occurred during OCR: {e}")
        return None

# Path to the image file
image_path = 'path_to_your_image.png'

# Perform OCR on the image
text_from_image = ocr_lithuanian(image_path)

# Print the recognized text
if text_from_image:
    print(text_from_image)
else:
    print("No text could be recognized.")
```
